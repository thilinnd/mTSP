```python
import pandas as pd
import requests
import numpy as np
from tqdm import tqdm
import time
```

```python

df = pd.read_csv("address.csv")
coordinates = list(zip(df["Latitude"], df["Longtitude"]))
labels = df["Address"].tolist()

# === 2. Cài đặt giới hạn (max điểm/lượt gọi)
MAX_POINTS = 100
n = len(coordinates)

# === 3. Hàm gọi OSRM giữa 2 nhóm điểm (nhóm nguồn và đích)
def get_osrm_matrix_chunk(src_coords, dst_coords):
    all_coords = src_coords + dst_coords
    coord_str = ";".join([f"{lon},{lat}" for lat, lon in all_coords])

    num_src = len(src_coords)
    num_dst = len(dst_coords)

    src_indices = ";".join(map(str, range(num_src)))
    dst_indices = ";".join(map(str, range(num_src, num_src + num_dst)))

    url = (
        f"http://localhost:5000/table/v1/driving/{coord_str}"
        f"?sources={src_indices}&destinations={dst_indices}&annotations=distance"
    )

    r = requests.get(url)
    if r.status_code != 200:
        raise Exception(f"OSRM error: {r.text}")
    return np.array(r.json()["distances"])

# === 4. Tạo ma trận rỗng
full_matrix = np.zeros((n, n))

# === 5. Chia thành các khối nhỏ (theo cả chiều hàng và cột)
for i in tqdm(range(0, n, MAX_POINTS)):
    for j in range(0, n, MAX_POINTS):
        src = coordinates[i:i + MAX_POINTS]
        dst = coordinates[j:j + MAX_POINTS]
        try:
            chunk = get_osrm_matrix_chunk(src, dst)
            full_matrix[i:i+len(src), j:j+len(dst)] = chunk
        except Exception as e:
            print(f"❌ Lỗi tại block ({i}, {j}): {e}")


# === 6. Chuyển sang km, lưu file
full_matrix_km = full_matrix / 1000
distance_df = pd.DataFrame(full_matrix_km, columns=labels, index=labels)
distance_df.to_csv("distance_matrix_km.csv", float_format="%.2f")

print("✅ Hoàn tất! Đã lưu file: distance_matrix_km.csv")
```
